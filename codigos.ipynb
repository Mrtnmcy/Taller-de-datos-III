graficar dispersion ancho de pata y peso en mathplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

surveys = pd.read_csv("surveys.csv")
my_plot = surveys.plot("hindfoot_length", "weight", kind="scatter")
plt.show()
```

graficar dispersion ancho de pata y peso por plotnine

```python
import plotnine as p9
myplot = (p9.ggplot(data=surveys,
                    mapping=p9.aes(x='hindfoot_length', y='weight')) + #variables
              p9.geom_point())
# converción de plotnine a matplotlib
my_plt_version = myplot.draw()

# Edicion 
p9_ax = my_plt_version.axes[0] 
p9_ax.set_xlabel("longitud pie trasero")
p9_ax.set_ylabel("peso")
p9_ax.tick_params(labelsize=16, pad=8)
p9_ax.set_title('Diagrama de peso versus longitud del pie trasero', fontsize=15)
plt.show()
```
